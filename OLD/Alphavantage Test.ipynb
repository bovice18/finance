```python
fullSnP = pd.read_csv('/home/bwood/Desktop/Fun/snp500-prices.csv')
```

```python
fullSnP.index = pd.to_datetime(fullSnP.index)
fullSnP = fullSnP.sort_index(ascending=True)
ytdSnP = fullSnP.loc['2020-1-1':'2020-4-21']
```

```python
ytdSnP.corr()
```

```python
snp = snp.set_index('Symbol')
```

```python
cleanPrices = ytdSnP.dropna(axis=1)
correlations = cleanPrices.corr()
```

```python
correlations = correlations['AAL'].to_frame().join(snp[['Security', 'GICS Sector', 'GICS Sub Industry']]).dropna()
```

```python
correlations.sort_values('AAL', ascending=False)
```

```python
### SHOW SOME SAMPLE RETURNS ###
returns = ytdSnP.apply(lambda x: (x / x[0]) - 1)
```

```python
### AIRLINES
airlineTickers = snp[snp['GICS Sub Industry'] == 'Airlines'].index.values

ax = returns[airlineTickers].plot()
# set monthly locator
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
# set formatter
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d-%Y'))
# set font and rotation for date tick labels
plt.gcf().autofmt_xdate()

plt.show()
```

```python
### FANG
fangTickers = ['FB', 'AMZN', 'NFLX', 'GOOGL']

ax = returns[fangTickers].plot()
# set monthly locator
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
# set formatter
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d-%Y'))
# set font and rotation for date tick labels
plt.gcf().autofmt_xdate()

plt.show()
```
